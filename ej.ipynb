```python
import matplotlib.pyplot as plt

name = ["Dany", "Ana", "Sara", "Juan"]
age = [30, 22, 27, 36]
colors_ = ['yellow', 'lightgrey', 'lightgreen','lightblue']


plt.bar(name,age, color = colors_)
plt.xlabel('nombre')
plt.ylabel('edad')
plt.title('Estudiantes')
plt.show()
```

```python
x = [2020, 2021, 2022, 2023]
y = [45, 46.4, 47, 45]
plt.plot(x,y, marker='o', linestyle='--', label='Ecuador', color= 'b')

x2 = [2020, 2021, 2022, 2023]
y2 = [45, 47, 48.6, 44]
plt.plot(x2,y2 ,marker='d', linestyle='-', label='Perú', color='r')



plt.xlabel('Año')
plt.ylabel('Población (millones)')
plt.yticks([45, 46, 47, 48, 44])
plt.xticks([2020,2021,2022,2023])
plt.legend(loc=('lower right'))
plt.show()


```
