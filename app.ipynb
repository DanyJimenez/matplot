```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Cambiar el tamaño de la figura
plt.figure(figsize=(6,4))
#lineplot: país 1
x = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
y = [40, 42, 43, 44, 45, 46, 48, 50]
plt.plot(x,y, marker = 'o', linestyle = '--', color ='y', label = 'Colombia')
#país 2
x2 = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
y2 = [41, 42, 43, 44, 44, 46, 50, 51]
plt.plot(x2,y2, marker = 'd', linestyle = '-', color ='r', label= 'Chile')
#agregar nombre a los ejes:
plt.xlabel('Años')
plt.ylabel('Población')
plt.title('Años vs Población')
plt.legend(loc = 'lower right') #ubicación de la leyenda
#mostrar solo los ejes que considero relevantes en la lista
#plt.yticks([50, 52, 54, 56])
plt.show()
```

```python
#subplots
fig, ax = plt.subplots(1,2,sharey=True)
ax[0].plot(x,y, color = 'y', label ='Colombia')
ax[0].legend()
ax[1].plot(x2, y2, color = 'r', label = 'Chile')
ax[1].legend()
plt.show()
```

```python
#barplot

x = ['Argentina', 'Brasil', 'Colombia']
y = [45, 90, 51]
colors_=['lightblue','greenyellow', 'yellow']

plt.bar(x,y, color=colors_)
plt.xlabel('country', fontsize= 15)
plt.ylabel('population (per million)', fontsize=15)
plt.show()
```

```python
#piecharts

pais = [ 'Chile', 'Perú', 'México']
poblacion = [33, 29, 120]

plt.pie (poblacion,labels =pais)
plt.show()
```

```python
#histograma:
edades = [22, 9, 15, 28, 30, 18, 41, 11]
bins = [12, 18, 25,35,41]
plt.hist(edades,bins, edgecolor='black')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de edades')
plt.show()
```

```python
#boxplots
plt.boxplot(edades)
plt.show()
```

```python
#scatterplot
a = [1, 2, 3, 4,7,8,10]
b = [2, 4, 7, 7, 9, 1, 3]
plt.scatter(a,b)
plt.show()
```
